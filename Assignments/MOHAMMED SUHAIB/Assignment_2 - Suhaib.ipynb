## **Finding the outliers and replacing the outliers**

```python
outliers = dataset.quantile(q=(0.3,0.85))
outliers
```

```python
outliers_removed = outliers.loc[0.85]-outliers.loc[0.3]
outliers_removed
```

```python
low_val = outliers.loc[0.3] - 1.5 * outliers_removed 
low_val
```

```python
high_val = outliers.loc[0.85] + 1.5 * outliers_removed
high_val
```

## **Checking for categorical data and encoding them**

```python
dataset.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
dataset['Gender'] = Encoder.fit_transform(dataset['Gender'])
dataset['Geography'] = Encoder.fit_transform(dataset['Geography'])
dataset['Surname'] = Encoder.fit_transform(dataset['Surname'])
```

```python
dataset.head()
```

## **Splitting the data into independent and dependent data**

```python
x = dataset.drop(columns=['Exited','RowNumber'],axis=1)
y = dataset['Exited']
```

```python
x.head()
```

## **Scaling the values**

```python
from sklearn.preprocessing import scale
X = scale(x)
X
```

```python
x = pd.DataFrame(X,columns = x.columns)
x
```

## **Splitting the data into training and testing data**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```
